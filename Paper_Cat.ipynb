```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
# Importing the fits file and converting it to a pandas dataframe

data = fits.open('Data/matched_catalogue.fits')
df = pd.DataFrame(data[1].data)
df
```

```python
ion_par = np.array(df.logq)
lum = np.array(df.HA6562_lum)
ew = np.array(df.EW_HA)
age = np.array(df.age)
```

```python
# Ionization Parameter vs Age of the HII regions

plt.plot(age, ion_par, 'o', markersize = 3)
plt.ylim(4, 12)
plt.xlim(0,25)
plt.xlabel('Age (Myr)')
plt.ylabel('Log Ionization Parameter')
plt.title('Age vs Ionization Parameter')
plt.show()

plt.loglog(age, ion_par, 'o', markersize = 3)
plt.xlabel('Age (Myr)')
# plt.xlim(0,25)
plt.ylabel('log Ionization Parameter')
plt.title('Ionization Parameter vs Age (log-log scale)') 
plt.show()
```

```python
# Equivalent Width of H-Î± vs Age of the HII regions 

plt.plot(age, ew, 'o', markersize = 3)
plt.xlabel('Age (Myr)')
plt.xlim(0,10)
plt.ylabel(r'Equivalent Width of H-Î± (m)')
plt.title(r'Age vs Equivalent Width of H-Î±')
plt.show()

plt.loglog(age, ew, 'o', markersize = 3)
plt.xlabel('Age (Myr)')
plt.xlim(0,25)
plt.ylabel(r'Equivalent Width of H-Î± (m)')
plt.title(r'Equivalent Width of H-Î± vs Age (log-log scale)')
plt.show()
```

```python
# Luminosity of H-Î± vs Age of the HII regions

plt.plot(age, lum, 'o', markersize = 3)
plt.xlabel('Age (Myr)')
plt.xlim(0,10)
# plt.ylim(-0.001, 0.001)
plt.ylabel(r'Luminosity of H-Î± (ergs)')
plt.title(r'Age vs Luminosity of H-Î±')
plt.show()

plt.loglog(age, lum, 'o', markersize = 3)
plt.xlabel('Age (Myr)')
plt.xlim(0,25)
plt.ylabel(r'Luminosity of H-Î± (ergs)')
plt.title(r'Luminosity of H-Î± vs Age (log-log scale)')
plt.show()
```

```python
# Sorting the dataframe by age, so that we can bin the data
df_sorted = df.sort_values(by = 'age')
df_sorted
```

```python
# Binning the data and calculating the median and mean of the ionization parameter, luminosity and equivalent width of H-Î±

bin_width = 5

ion_meds = []
lum_meds = []
ew_meds = []

ion_means = []
lum_means = []
ew_means = []

bins = np.arange(int(min(df_sorted['age'])), int(max(df_sorted['age'])), bin_width)

for i in bins:
    ion_bin = np.array(list(df_sorted['logq'][df_sorted['age'] > i][df_sorted['age'] < i + bin_width]))
    lum_bin = np.array(list(df_sorted['HA6562_lum'][df_sorted['age'] > i][df_sorted['age'] < i + bin_width]))
    ew_bin = np.array(list(df_sorted['EW_HA'][df_sorted['age'] > i][df_sorted['age'] < i + bin_width]))

    if ion_bin.size == 0:
        ion_meds.append(np.nan)
        ion_means.append(np.nan)
    else:
        ion_meds.append(np.nanmedian(ion_bin))
        ion_means.append(np.nanmean(ion_bin))


    if lum_bin.size == 0:
        lum_meds.append(np.nan)
        lum_means.append(np.nan)
    else:
        lum_meds.append(np.nanmedian(lum_bin))
        lum_means.append(np.nanmean(lum_bin))

    if ew_bin.size == 0:
        ew_meds.append(np.nan)
        ew_means.append(np.nan)
    else:
        ew_meds.append(np.nanmedian(ew_bin))
        ew_means.append(np.nanmean(ew_bin))

        
    
```

```python
# Plotting the binned ionization parameter data

plt.plot(bins, ion_meds, 'o', markersize = 5)
plt.xlabel('Binned Age (Myr) with bin width = ' + str(bin_width))
plt.ylabel('Median Ionization Parameter')
plt.xlim(0, 40)
plt.title('Binned Age vs Median Ionization Parameter')
plt.show()

plt.plot(bins, ion_means, 'o', markersize = 5)
plt.xlabel('Binned Age (Myr) with bin width = ' + str(bin_width))
plt.ylabel('Mean Ionization Parameter')
plt.xlim(0, 40)
plt.title('Binned Age vs Mean Ionization Parameter')
plt.show()
```

```python
# Plotting the binned equivalent width data

plt.plot(bins, ew_meds, 'o', markersize = 5)
plt.xlabel('Binned Age (Myr) with bin width = ' + str(bin_width))
plt.ylabel('Median Equivalent Width of H-Î± (m)')
plt.xlim(0, 40)
plt.title('Binned Age vs Median Equivalent Width of H-Î±')
plt.show()

plt.plot(bins, ew_means, 'o', markersize = 5)
plt.xlabel('Binned Age (Myr) with bin width = ' + str(bin_width))
plt.ylabel('Mean Equivalent Width of H-Î± (m)')
plt.xlim(0, 40)
plt.title('Binned Age vs Mean Equivalent Width of H-Î±')
plt.show()
```

```python
# Plotting the binned luminosity data

plt.plot(bins, lum_meds, 'o', markersize = 3)
plt.xlabel('Binned Age (Myr) with bin width = ' + str(bin_width))
plt.ylabel('Median Luminosity of H-Î± (ergs)')
plt.xlim(0, 40)
plt.title('Binned Age vs Median Luminosity of H-Î±')
plt.show()

plt.plot(bins, lum_means, 'o', markersize = 3)
plt.xlabel('Binned Age (Myr) with bin width = ' + str(bin_width))
plt.ylabel('Mean Luminosity of H-Î± (ergs)')
plt.xlim(0, 25)
plt.title('Binned Age vs Mean Luminosity of H-Î±')
plt.show()
```
