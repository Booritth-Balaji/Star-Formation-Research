```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the cross matched catalogue of the MUSE and UVIT data which contains the estimated ages of HII regions in the galaxy NGC 1566
# 1.5 arcsec error was used while cross matching the data

data = pd.read_csv('Data/Cross Matched MUSE and UVIT with Ages', delimiter = ',')
data
```

```python
# Loading the data which was cross matched with 2 arc seconds error
data_2 = pd.read_csv('Data/Cross Matched MUSE and UVIT with 2 arcsec', delimiter= ',')
data_2
```

```python
# We are interested in the following properties: 

# logqD91 (Logarithm of the Ionization parameter q)
# FHa6562corr (Integrated Flux of the Halpha 6562Å emission line corrected for dust extinction) in 10^-23 ergs/m^2
# EW_HA6562_fit (Equivalent width of H-α, measured after stellar continuum subtracted) in Angstroms
# clump_age (Age of the HII region) in Myr

ion_par = np.array(data.logqD91)
lum = np.array(data.FHa6562corr) * 4 * np.pi * (17.69* 10**6 * 3.086 * 10**16)**2 * 10**(-23) * 10**7 # in ergs # 17.69 Mpc is the distance to the galaxy NGC 1566
ew = np.array(data.EW_HA6562_fit) * 10**(-10) # in m
age = np.array(data.clump_age) # in Myr

ion_par_2 = np.array(data_2.logqD91)
lum_2 = np.array(data_2.FHa6562corr) * 4 * np.pi * (17.69* 10**6 * 3.086 * 10**16)**2 * 10**(-23) * 10**7 # in ergs
ew_2 = np.array(data_2.EW_HA6562_fit) * 10**(-10) # in m
age_2 = np.array(data_2.clump_age) # in Myr
```

```python
# Ionization Parameter vs Age of the HII regions

plt.scatter(age, ion_par,  label = '1.5 arcsec error')
plt.ylim(6, 9)
plt.xlim(0,40)
plt.xlabel('Age (Myr)')
plt.ylabel('Log Ionization Parameter')
plt.title('Ionization Parameter vs Age')
# plt.scatter(age_2, ion_par_2,  label = '2 arcsec error', marker = 'x')
plt.legend()
plt.show()


# plt.loglog(age, ion_par, 'o', markersize = 3)
# plt.xlabel('Age (Myr)')
# plt.xlim(0,25)
# plt.ylabel('log Ionization Parameter')
# plt.title('Ionization Parameter vs Age (log-log scale)') 
# plt.show()
```

```python
# Equivalent Width of H-α vs Age of the HII regions 

plt.scatter(age, ew,  label = '1.5 arcsec error')
# plt.scatter(age_2, ew_2,  label = '2 arcsec error', marker = 'x')
plt.xlim(0,40)
plt.xlabel('Age (Myr)')
plt.ylabel(r'Equivalent Width of H-α (m)')
# plt.legend()
plt.title(r'Equivalent Width of H-α vs Age')
plt.show()

# plt.loglog(age, ew, 'o', markersize = 3)
# plt.xlabel('Age (Myr)')
# plt.xlim(0,25)
# plt.ylabel(r'Equivalent Width of H-α (m)')
# plt.title(r'Equivalent Width of H-α vs Age (log-log scale)')
# plt.show()
```

```python
# Luminosity of H-α vs Age of the HII regions

plt.scatter(age, lum,  label = '1.5 arcsec error')
# plt.scatter(age_2, lum_2,  label = '2 arcsec error', marker = 'x')
plt.xlabel('Age (Myr)')
plt.xlim(0,40)
plt.ylim(-0.2e40, 2e40)
plt.ylabel(r'Luminosity of H-α (ergs)')
# plt.legend()
plt.title(r'Luminosity of H-α vs Age')
plt.show()

plt.loglog(age, lum, 'o', markersize = 3)
plt.xlabel('Age (Myr)') 
plt.xlim(0,25)
plt.ylabel(r'Luminosity of H-α (ergs)')
plt.title(r'Luminosity of H-α vs Age (log-log scale)')
plt.show()
```

```python
# Binning the data and calculating the median and mean of the ionization parameter, luminosity and equivalent width of H-α

bin_width = 5

ion_meds = []
lum_meds = []
ew_meds = []

ion_means = []
lum_means = []
ew_means = []

bins = np.arange(int(min(data['clump_age'])), int(max(data['clump_age'])), bin_width)

for i in bins:
    ion_bin = np.array(list(data['logqD91'][data['clump_age'] > i][data['clump_age'] < i + bin_width]))
    lum_bin = np.array(list(data['FHa6562corr'][data['clump_age'] > i][data['clump_age'] < i + bin_width]))
    lum_bin = lum_bin * 4 * np.pi * (17.69* 10**6 * 3.086 * 10**16)**2 * 10**(-23) * 10**7
    ew_bin = np.array(list(data['EW_HA6562_fit'][data['clump_age'] > i][data['clump_age'] < i + bin_width]))
    ew_bin = ew_bin * 10**(-10)
    

    if ion_bin.size == 0:
        ion_meds.append(np.nan)
        ion_means.append(np.nan)
    else:
        ion_meds.append(np.nanmedian(ion_bin))
        ion_means.append(np.nanmean(ion_bin))


    if lum_bin.size == 0:
        lum_meds.append(np.nan)
        lum_means.append(np.nan)
    else:
        lum_meds.append(np.nanmedian(lum_bin))
        lum_means.append(np.nanmean(lum_bin))

    if ew_bin.size == 0:
        ew_meds.append(np.nan)
        ew_means.append(np.nan)
    else:
        ew_meds.append(np.nanmedian(ew_bin))
        ew_means.append(np.nanmean(ew_bin))

        
    
```

```python
# Plotting the binned ionization parameter data

plt.plot(bins, ion_meds, 'o', markersize = 5)
plt.xlabel('Binned Age (Myr) with bin width = ' + str(bin_width))
plt.ylabel('Median Ionization Parameter')
plt.xlim(0, 40)
plt.title('Binned Age vs Median Ionization Parameter')
plt.show()

plt.plot(bins, ion_means, 'o', markersize = 5)
plt.xlabel('Binned Age (Myr) with bin width = ' + str(bin_width))
plt.ylabel('Mean Ionization Parameter')
plt.xlim(0, 40)
plt.title('Binned Age vs Mean Ionization Parameter')
plt.show()
```

```python
# Plotting the binned equivalent width data

plt.plot(bins, ew_meds, 'o', markersize = 5)
plt.xlabel('Binned Age (Myr) with bin width = ' + str(bin_width))
plt.ylabel('Median Equivalent Width of H-α (m)')
plt.xlim(0, 40)
plt.title('Binned Age vs Median Equivalent Width of H-α')
plt.show()

plt.plot(bins, ew_means, 'o', markersize = 5)
plt.xlabel('Binned Age (Myr) with bin width = ' + str(bin_width))
plt.ylabel('Mean Equivalent Width of H-α (m)')
plt.xlim(0, 40)
plt.title('Binned Age vs Mean Equivalent Width of H-α')
plt.show()
```

```python
# Plotting the binned luminosity data

plt.plot(bins, lum_meds, 'o', markersize = 3)
plt.xlabel('Binned Age (Myr) with bin width = ' + str(bin_width))
plt.ylabel('Median Luminosity of H-α (ergs)')
plt.xlim(0, 40)
plt.ylim(-0.2e40, 0.4e40)
plt.title('Binned Age vs Median Luminosity of H-α')
plt.show()

plt.plot(bins, lum_means, 'o', markersize = 3)
plt.xlabel('Binned Age (Myr) with bin width = ' + str(bin_width))
plt.ylabel('Mean Luminosity of H-α (ergs)')
plt.xlim(0, 25)
plt.title('Binned Age vs Mean Luminosity of H-α')
plt.show()
```
